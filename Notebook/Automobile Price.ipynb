```python
car_v1 = car.replace({'engine-type': {'l':1, 'ohc':2, 'rotor':3, 'dohc':4, 'ohcv':5, 'ohcf':6},
                      'num-of-cylinders': {'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8},
                      'body-style': {'hatchback':1, 'wagon':2, 'sedan':3, 'convertible':4, 'hardtop':5}})
```

```python
car_v1.head()
```

```python
car_v1 = pd.get_dummies(car_v1)
```

```python
car_v1.info()
```

```python
car_v1.corr()
```

```python
car_v1.corr().style.background_gradient()
```

```python
plt.figure(figsize=(10,12))
sns.set_style('white')
sns.heatmap(data=car_v1[['wheel-base', 'length', 'engine-type', 'num-of-cylinders',
                      'horsepower', 'average-mileage', 'price']].corr(), linewidths=0.5, cmap='coolwarm')
plt.suptitle('Correlation Matrix of Cars')
plt.show()
```

```python
feature = car_v1.drop('price', axis=1)
target = car_v1['price']
```

```python
car.head()
```

```python
target.head()
```

```python
feature.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature, target, test_size=0.30, random_state=0)
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
car_model = LinearRegression() #define the model
car_model.fit(x_train, y_train) #fitting the model in traing data
```

```python
m = car_model.coef_             # checking m and c
c = car_model.intercept_
```

```python
m, c
```

```python
y_pred_test = car_model.predict(x_test)
y_pred_test
```

```python
y_test
```

```python
y_prediction_train = car_model.predict(x_train)
```

```python
y_prediction_train
```

```python
y_train.head(10)
```

```python
y_train.tail(10)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred_test) #model validation
```

```python
accuracy = r2_score(y_test, y_pred_test)
print('The model is accurate of', accuracy*100)
```

```python
print(r2_score(y_train, y_prediction_train))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))

sns.regplot(x=y_test, y=y_pred_test, ci=None, color='b')
```

```python
plt.figure(figsize=(4, 4))
plt.style.use("ggplot")

plt.scatter(x=range(10), y=y_test.head(10), color='r', label='Actual')
plt.scatter(x=range(10), y=y_pred_test[:10], color='b', label='Predicted')
plt.legend()
```
